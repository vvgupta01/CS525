```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_theme(style="white")
```

## Simulation

```python
sim = pd.read_csv("sim.csv", 
                  names=["alloc", "sigma", "avg_utilization", "avg_welfare", "welfare_disparity", "fairness", "avg_fairness"])
sim.head()
```

### Utilization

```python
plt.figure(figsize=(6, 4))

sns.lineplot(x="sigma", y="avg_utilization", hue="alloc", data=sim, marker="o")
plt.legend(title=None, loc="lower right")
plt.xlabel("Selfish Users (%)")
plt.ylabel("Average Utilization")

plt.savefig("utilization.png", bbox_inches="tight")
plt.show()
```

### Welfare

```python
plt.figure(figsize=(6, 8))

plt.subplot(2, 1, 1)
sns.lineplot(x="sigma", y="avg_welfare", hue="alloc", data=sim, marker="o")
plt.legend(title=None, loc="upper right")
plt.xlabel("Selfish Users (%)")
plt.ylabel("Average Welfare")

plt.subplot(2, 1, 2)
sns.lineplot(x="sigma", y="welfare_disparity", hue="alloc", data=sim, marker="o")
plt.legend(title=None, loc="upper right")
plt.xlabel("Selfish Users (%)")
plt.ylabel("Altruism Incentive")

plt.savefig("welfare.png", bbox_inches="tight")
plt.show()
```

### Fairness

```python
plt.figure(figsize=(6, 8))

plt.subplot(2, 1, 1)
sns.lineplot(x="sigma", y="fairness", hue="alloc", data=sim, marker="o")
plt.legend(title=None, loc="upper right")
plt.xlabel("Selfish Users (%)")
plt.ylabel("Long-term Fairness")

plt.subplot(2, 1, 2)
sns.lineplot(x="sigma", y="avg_fairness", hue="alloc", data=sim, marker="o")
plt.legend(title=None, loc="upper right")
plt.xlabel("Selfish Users (%)")
plt.ylabel("Average Short-term Fairness")

plt.savefig("fairness.png", bbox_inches="tight")
plt.show()
```

## Preliminary Benchmark

```python
bench = pd.read_csv("bench.csv", names=["alloc", "num_tenants", "avg_runtime"])
bench.head()
```

### Runtime

```python
plt.figure(figsize=(6, 4))

sns.lineplot(x="num_tenants", y="avg_runtime", hue="alloc", data=bench, marker="o").set(xscale="log")
plt.legend(title=None, loc="upper left")
plt.xlabel("Number of Users")
plt.ylabel("Runtime (ms)")

plt.savefig("runtime.png", bbox_inches="tight")
plt.show()
```
